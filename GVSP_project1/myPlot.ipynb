```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("/home/g-v-sai-praneeth/App_Stats/archive/GaltonFamilies.csv")
```

```python
data.head()
```

```python
Father_ht=data[data['childNum']==1]['father']
Mother_ht=data[data['childNum']==1]['mother']
Child_ht=data['childHeight']
```

```python
plt.hist(Father_ht,bins=10,color='skyblue', edgecolor='black')
plt.title("Distribution of Father's height")
plt.xlabel("Height in inches")
plt.ylabel("Frequency")
plt.savefig("Father_height.png", dpi=300)  # Higher DPI for better quality

plt.show()
```

```python
plt.hist(Mother_ht,bins=7,color='skyblue', edgecolor='black')
plt.title("Distribution of Mother's height")
plt.xlabel("Height in inches")
plt.ylabel("Frequency")
plt.savefig("Mother_height.png", dpi=300)  # Higher DPI for better quality

plt.show()
```

```python
plt.hist(Child_ht,bins=11,color='skyblue', edgecolor='black')
plt.title("Distribution of Children's height")
plt.xlabel("Height in inches")
plt.ylabel("Frequency")
plt.savefig("Children_height.png", dpi=300)  # Higher DPI for better quality

plt.show()
```

```python
plt.boxplot([Father_ht,Mother_ht,Child_ht],labels=['Father','Mother','Child'],patch_artist=True,boxprops=dict(facecolor='skyblue',color='blue'))
plt.savefig("box_plot",dpi=300)
plt.show()
```

```python
Father_bins=[0,65,70,75,float('inf')]
Father_labels=['<65','65-70','70-75','>75']
Father_pie=pd.cut(Father_ht,bins=Father_bins,labels=Father_labels)
Father_count=Father_pie.value_counts().sort_index()
plt.figure(figsize=(7,7))
plt.pie(Father_count,labels=Father_count.index,autopct='%0.1f%%',colors=plt.cm.Paired.colors,startangle=140)
plt.title("Height distribution of Father's")
plt.savefig("PieFather",dpi=300)
plt.show()
```

```python
Mother_bins=[0,60,62,64,66,68,float('inf')]
Mother_labels=['<60','60-62','62-64','64-66','66-68','>68']
Mother_pie=pd.cut(Mother_ht,bins=Mother_bins,labels=Mother_labels)
Mother_count=Mother_pie.value_counts().sort_index()
plt.figure(figsize=(7,7))
plt.pie(Mother_count,labels=Mother_count.index,autopct='%0.1f%%',colors=plt.cm.Paired.colors,startangle=140)
plt.title("Height distribution of Mother's")
plt.savefig("PieMother",dpi=300)
plt.show()
```

```python
Child_bins=[0,60,65,70,75,float('inf')]
Child_labels=['<60','60-65','65-70','70-75','>75']
Child_pie=pd.cut(Child_ht,bins=Child_bins,labels=Child_labels)
Child_count=Child_pie.value_counts().sort_index()
plt.figure(figsize=(7,7))
plt.pie(Child_count,labels=Child_count.index,autopct='%0.1f%%',colors=plt.cm.Paired.colors,startangle=140)
plt.title("Height distribution of Children's")
plt.savefig("PieChildren",dpi=300)
plt.show()
```

```python
def scatter_plot(s):
    corr_coeff=data[s].corr(data['childHeight'])
    print(f"correlation_coefficient between {s}'s height and child height:{corr_coeff:0.4f}")
    plt.scatter(data[s],data['childHeight'],alpha=0.5)
    plt.title(f"{s} vs child height")
    plt.xlabel(f"{s} height")
    plt.ylabel(f"child height")
    plt.grid(True)
    plt.savefig(f"{s} vs child scatter plot",dpi=300)
    plt.show()
scatter_plot("father")
scatter_plot("mother")
scatter_plot("midparentHeight")
```
